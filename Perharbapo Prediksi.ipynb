```python
test = data.iloc[:, 0].values
test_1 = data.iloc[:, 1].values
test_2 = data.iloc[:, 2].values
test_3 = data.iloc[:, 3].values
test_4 = data.iloc[:, 4].values
train = data.iloc[:, 1:].values
train_1 = data.iloc[:, [0,2,3,4,5,6,7]].values
train_2 = data.iloc[:, [0,1,3,4,5,6,7]].values
train_3 = data.iloc[:, [0,1,2,4,5,6,7]].values
train_4 = data.iloc[:, [0,1,2,3,5,6,7]].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.1, random_state=19)
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(train_1, test_1, test_size=0.1, random_state=19)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(train_2, test_2, test_size=0.1, random_state=19)
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(train_3, test_3, test_size=0.1, random_state=19)
X_train_4, X_test_4, y_train_4, y_test_4 = train_test_split(train_4, test_4, test_size=0.1, random_state=19)
```

```python
svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
svr.score(X_test, y_test)
```
